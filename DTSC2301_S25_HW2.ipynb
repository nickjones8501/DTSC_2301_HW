#### Question 1
In the file `https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ironman1819.csv', there is data on female finishers of the 2018 and 2019 Ironman Triathlon in Lake Placid, NY.  We will focus on the times to complete the three elements of the triathlon, swimming (Swim.Time), biking (Bike.Time) and running (Run.Time).  The units for each of these features is minutes.  The overall time (including transitions from on element to the next) is found in *Overall.Time*. Create a scatterplot for *Swim.Time* as a predictor for *Run.Time*.  As a note it is generally not a good idea in python to make feature/variables have a '.' in them.

More on these data as well as a video introduction to them can be found at: https://isle.stat.cmu.edu/SCORE/ironman_triathlon/

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load file
file_path = 'https://webpages.charlotte.edu/mschuck1/classes/DTSC2301/Data/Ironman1819.csv'  # Adjust this path if needed
data = pd.read_csv(file_path)

# Initialize variables
plt.scatter(data['Swim.Time'], data['Run.Time'], color="blue")

# Add labels and title
plt.xlabel('Swim Time (minutes)')
plt.ylabel('Run Time (minutes)')
plt.title('Scatterplot of Swim Time vs Run Time')

# Show the plot
plt.show()
```

#### Question 2
Using the Ironman data from above, fit a regression model that predicts *Overall.Time* from *Bike.Time* and add that line to a scatterplot of those two variables.  Interpret the slope and the y-intercept for these data in the context of these data.  

```python
# Scatterplot of Bike.Time vs Overall.Time
plt.scatter(data['Bike.Time'], data['Overall.Time'], color="blue")

# Add labels and title
plt.xlabel('Bike Time (minutes)')
plt.ylabel('Overall Time (minutes)')
plt.title('Regression: Overall Time vs Bike Time')

# Fit the regression model
# Calculate slope (beta_1) and intercept (beta_0)
beta_1, beta_0 = np.polyfit(data['Bike.Time'], data['Overall.Time'], deg=1)

# Create an array of x values for the regression line
xseq = np.linspace(data['Bike.Time'].min(), data['Bike.Time'].max(), num=100)

# Add regression line to plot
plt.plot(xseq, beta_0 + beta_1 * xseq, color='red')

# Set the x-axis and the y-axis limits (optional, adjust as needed)
plt.xlim(data['Bike.Time'].min() - 10, data['Bike.Time'].max() + 10)
plt.ylim(data['Overall.Time'].min() - 50, data['Overall.Time'].max() + 50)

# Show plot
plt.show()

# Print the slope and intercept
print("Intercept (beta_0):", beta_0)
print("Slope (beta_1):", beta_1)
```
